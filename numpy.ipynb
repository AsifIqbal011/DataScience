```python
A=np.array([[1,2],[3,4]])
B=np.array([[2,4],[6,8]])
print("sum:")
print(A+B)
print("sub:")
print(A-B)
print("multi:")   #elementwise product
print(A*B)
print("Division:")
print(A/B)
```

```python
A@B #matix product
```

```python
A.dot(B)  # another matrix product
```

```python
arr1=np.arange(12).reshape(3,4)
arr1
```

```python
print(arr1.sum())     #axis=0 -> column
print(arr1.min())     #axis=1 -> row
print(arr1.max())

print(arr1.sum(axis=0)) # column wise sum
print()
print(arr1.sum(axis=1)) #row wise sum
print()
print(arr1.mean(axis=1))
```
