Merging & Concatination

```python
customers = pd.DataFrame({
    "Customer_ID": [1, 2, 3, 4],
    "Name": ["Rahim", "Karim", "Asif", "Nusrat"],
    "City": ["Dhaka", "Chattogram", "Dhaka", "Sylhet"]
})

customers
```

```python
orders = pd.DataFrame({
    "Order_ID": [101, 102, 103, 104, 105, 106],
    "Customer_ID": [1, 2, 2, 6, 3, 5],
    "Product": ["Fan", "LED Light", "Heater","Bulb", "Fan", "Rice Cooker"],
    "Amount": [3500, 800, 4500,2100, 3500, 4200]
})

orders
```

```python
pd.merge(customers,orders,on='Customer_ID' ,how='inner') # default its inner join
```

```python
statement=pd.merge(customers,orders,on='Customer_ID', how='outer')
statement
```

```python
statement.isnull().sum()
```

```python
pd.merge(customers,orders,on='Customer_ID', how='left') #Keep all customer
```

```python
pd.merge(customers,orders,on='Customer_ID', how='right') #Keep all orders
```

```python
pd.merge(customers,orders, how='cross') #cross join
```

```python
orders_part2 = pd.DataFrame({
    "Order_ID": [106, 107],
    "Customer_ID": [1, 4],
    "Product": ["Heater", "Fan"],
    "Amount": [4500, 3500]
})
orders_part2
```

```python
pd.concat([orders,orders_part2],ignore_index=True)
```

```python
pd.concat([orders,orders_part2],axis=1) #column-wise concatination
```

```python
extra_info = pd.DataFrame({
    "Membership": ["Gold", "Silver", "Gold", "Bronze"]
})
extra_info
```
